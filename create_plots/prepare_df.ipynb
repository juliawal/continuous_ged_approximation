```
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
mao = pd.read_csv('/content/drive/MyDrive/NS_Projekt/mao_all.csv')
rna = pd.read_csv('/content/drive/MyDrive/NS_Projekt/rna_all.csv')
acyclic = pd.read_csv('/content/drive/MyDrive/NS_Projekt/acyclic_all.csv')
```

```
col = sns.color_palette("mako")
```

```
def diff(a,b):
    return b-a


def lower(a,b):
  if a < b:
    return 1
  if b < a:
    return -1
  else:
    return 0
```

```
mao['Dataset']='MAO'
rna['Dataset']='RNA'
acyclic['Dataset']='Acyclic'

mao['Runtime Hungarian'] = np.mean(mao['Runtime Hungarian'])
rna['Runtime Hungarian'] = np.mean(rna['Runtime Hungarian'])
acyclic['Runtime Hungarian'] = np.mean(acyclic['Runtime Hungarian'])
mao['Mean Runtime Sinkhorn'] = np.mean(mao['Runtime Sinkhorn'])
rna['Mean Runtime Sinkhorn'] = np.mean(rna['Runtime Sinkhorn'])
acyclic['Mean Runtime Sinkhorn'] = np.mean(acyclic['Runtime Sinkhorn'])
```

```
df = pd.concat([acyclic, mao, rna])
```

```
df['Induced Edit Cost Sinkhorn Lower'] = df.apply(lambda x: lower(x['Induced Edit Cost Sinkhorn'], x['Induced Edit Cost Hungarian']), axis=1)
```

```
df = df = df.rename({'Relative Error Upper Bound': 'Error Induced Edit Cost', 'T_max': 'alpha', 'Runtime Hungarian': 'Mean Runtime Hungarian','T_param': 'beta', 'Relative Diversion Matching Cost Sinkhorn Projection': 'Diversion Matching Cost Sinkhorn Projection', 'Relative Diversion Matching Cost Sinkhorn': 'Diversion Matching Cost Sinkhorn'}, axis=1)
df = df.drop('Unnamed: 0', axis=1)
```
